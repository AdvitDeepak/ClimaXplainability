```python
import climate_learn as cl
import matplotlib.pyplot as plt
import numpy as np
import os
# temperature = np.array([
#     [0,0,0,1,1,1,2,2,2],
#     [0,0,0,1,1,1,2,2,2],
#     [0,0,0,1,1,1,2,2,2],
# ])


path = '/home/advit/ClimateData/run_dumps/'
i = 0
for file in os.listdir(path)[::-1]:
    file_path = os.path.join(path, file)
    temperature = np.load(file_path).squeeze()

    # Create a heatmap
    plt.figure(figsize=(temperature.shape[1], temperature.shape[0]))
    heatmap = plt.pcolor(temperature, cmap='bwr')
    plt.colorbar(heatmap)

    # Set labels for the axes (optional)
    plt.xticks(np.arange(temperature.shape[1]) + 0.5, range(temperature.shape[1]))
    plt.yticks(np.arange(temperature.shape[0]) + 0.5, range(temperature.shape[0]))

    plt.xlabel('width of the earth??')
    plt.ylabel('height of the earth??')
    plt.title(file, fontsize=100)
    plt.show()

    print(np.min(temperature))
    print(np.max(temperature))
    print(np.mean(temperature))

    i += 1
    if i == 4:
        break
```
